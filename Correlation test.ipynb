# Correlation test

Pearson's correlation assumes both variables are continuous and normally distributed!
In summary, while Pearson's correlation can be used for encoded categorical columns under certain conditions (specifically, ordinal data with an encoding that reflects its natural order). we can make a plot and only look at the ordinal columns.

Chi-Square Test of Independence: This is a common method to test the relationship between two categorical variables. It checks whether the distribution of sample categorical data matches an expected distribution.


Spearman's correlation is ideal for ordinal data!

# Spearman Correlation

Unlike Pearson's correlation, does not measure linear correlation directly. Instead, it assesses how well the relationship between two variables can be described using a monotonic function. In simpler terms, it measures the strength and direction of a monotonic relationship between two variables.



```python
import pandas as pd
from scipy.stats import spearmanr

df = df.dropna()#Notice spearman correlation can't deal with nas, we have to drop it.
spearman_corr, p_value = spearmanr(df['Household Income'], df['How often do you travel by plane?'])


print(f"Spearman Correlation Coefficient: {spearman_corr}")
print(f"P-value: {p_value}")
# Interpret the results
if p_value < 0.05:  # Common threshold for statistical significance
    print("There is a significant association between Household Income and Education.")
else:
    print("There is no significant association between Household Income and Education.")

```

# Pearson Correlation plot

It measures the linear correlation between pairs of attributes in the dataframe.
The Pearson correlation coefficient ranges from -1 to 1, where:

1 implies a perfect positive linear relationship,
-1 implies a perfect negative linear relationship, and
0 implies no linear correlation.



```python
df=pd.read_csv('/Users/xinqiaoyang/Desktop/coded.csv',sep=',') #read the file again because earlier we dropped the nas.
corr = df.corr(numeric_only=True)
corr.style.background_gradient(cmap='coolwarm') #we can make a plot and only look at the ordinal columns!
```
