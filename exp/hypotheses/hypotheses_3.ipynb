```python
import sys
sys.path.append('../../src')

import pandas as pd
from tueplots.constants.color import rgb
from create_plots_feature import MakeFeaturePlots

PATH_CODED_CSV = '../../data/coded.csv'
```

```python
%%capture --no-display

import matplotlib.pyplot as plt
from tueplots import bundles
import os


condition = 'How tall are you?'
event = "Given the opportunity, would you eliminate the possibility of reclining seats on planes entirely?"


df  = pd.read_csv(PATH_CODED_CSV)

probabilities = df.groupby(condition)[event].value_counts(normalize=True).unstack()

probabilities.plot(kind='line', stacked=False, color=['#ff0000', '#0000ff'], linewidth=2)
plt.rcParams.update(bundles.icml2022(column='half', nrows=1, ncols=1, usetex=False))
plt.rcParams.update({
    'text.usetex': False,
    'font.family': 'serif',
    'font.serif': 'Times New Roman',
    'text.latex.preamble': '\\usepackage{times} ',
    'figure.figsize': (3.25*2.77, 2.0086104634371584*1.4),
    'figure.constrained_layout.use': True,
    'figure.autolayout': False,
    'savefig.bbox': 'tight',
    'savefig.pad_inches': 0.015,
    'font.size': 11+7,
    'axes.labelsize': 8+5,
    'legend.fontsize': 8+5,
    'xtick.labelsize': 8+5,
    'ytick.labelsize': 8+5,
    'axes.titlesize': 11+7
})

# Move legend outside the plot

plt.legend(loc='upper right', labels=['No','Yes'])
plt.title("Given the opportunity, would you eliminate the possibility of \n reclining seats on planes entirely?")
path_fig = '../../doc/AnalysisPassengerBehaviour/fig'
plt.savefig(os.path.join(path_fig, "hypo_3.pdf"))
plt.show()
```
